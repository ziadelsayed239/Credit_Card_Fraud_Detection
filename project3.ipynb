# Handling Imbalanced Data

```python
x=df.drop(['Class',"V2","V3","V5"],axis=1)
y=df['Class']
x_train , x_test , y_train , y_test =train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(k_neighbors=7)
x_train_sm,y_train_sm=sm.fit_resample(x_train,y_train)
```

```python
data_after_smote = pd.DataFrame(x_train_sm, columns=x_train.columns)
data_after_smote['target'] = y_train_sm 
```

```python
data_after_smote.shape
```

```python
sns.countplot(data_after_smote["target"])
plt.show()
```

# Split Data

```python
X= data_after_smote.drop(['target'],axis=1)
Y=data_after_smote['target']
```

# Scailing Data

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

# Modeling

### Random Forest

```python
RF = RandomForestClassifier(n_estimators=100, random_state=42)  
RF.fit(X_train, y_train)
```

```python
RF.score(X_train,y_train)
```

```python
RF.score(X_test,y_test)
```

```python
y_pred=RF.predict(X_test)
y_pred
```

```python
df=pd.DataFrame({"y_predect":y_pred,"y_test":y_test})
df
```

```python
report = classification_report(y_test, y_pred)
print(report)
```
